^ 오류가 나는 이유 - 스택에 저장하는 게 무한정할 수 없기 때문에.

```python
# code 10-02. 재귀 호출 함수 기본(반환 조건 추가)

def openBox():
    global count
    print("종이 상자를 엽니다. ^^")
    count -= 1
    if count == 0:
        print("** 반지를 넣고 반환합니다. **")
        return
    openBox()    
    print("종이 상자를 닫습니다.")



count = 5
openBox()
```

```python
# 숫자 합계 내기 : for 문

sumValue = 0
for n in range(10,0,-1):
    sumValue += n

print("10+9+.....+1 = ", sumValue)
```

```python
# code 10-03 숫자 합계 내기 : 재귀

def addNumber(num):
    if num <= 1:
        return 1
    return num + addNumber(num-1)


print(addNumber(5))
```

```python
# 숫자 곱셈하기 : for 문

factValue = 1
for n in range(10, 0, -1):
    factValue *= n

print("10*9*.....*1 = ", factValue)
```

```python
# 숫자 곱셈하기 : 재귀

def factorial(num):
    if num <= 1:
        return 1
    return num * factorial(num-1)

print("\n10! = ", factorial(10))
```
